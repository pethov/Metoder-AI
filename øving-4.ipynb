### i
Struktur: S->F<-W
Dette betyr:  $S \perp W$ som ikke er riktig i forhold til historen så det kan ikke være denne.

### ii

struktur: W->S->F<-W
Dette betyr $S \not\!\perp W \mid F$ som  ikke er riktig i forhold til hisorien så kan ikke være denne.

### iii

struktur: S<-F->W
Dette betyr $S \perp W \mid F$. som stemmer overens med det vi tolket utifra teksten.

Vi ønsker også å match kausal struktur fra historen. Som betyr retningene på pilene samsvarer med hvordan prossesen faktisk skjer i historen. som bare kan være nettverk iii

Siden EU_safe>EU_lotteri så er det rasjonale valget å velge å ta alternativ 1 altså det safe valget.

```python
import numpy as np
from scipy.optimize import fsolve

# Likningen vi vil løse:
# U(100) = 0.5 * U(500) + 0.5*U(0)

def equation(R):
    return U(100, R) - (0.5 * U(500, R) + 0.5 * U(0, R))

# Startgjetning
R_initial_guess = 100

R_solution = fsolve(equation, R_initial_guess)

print("R ≈", round(R_solution[0], 0))
```
